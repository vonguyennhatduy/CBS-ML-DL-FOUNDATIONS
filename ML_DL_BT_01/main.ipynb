# Tính MSE cho dữ liệu giá xe ô tô

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

### Data

```python
car_age = np.array([1, 2, 3, 4, 5, 6])
actual_price  = np.array ([20000, 18000, 15000, 12000, 10000, 8000])
predicted_price  = np.array ([21000, 17500, 15500, 11500, 9500, 8500])
```

### Tính MSE sử dụng thư viện scikit-learn

```python
mse_car = mean_squared_error(actual_price, predicted_price)
print("Mean Squared Error (MSE):", mse_car)
```

### Tính MSE sử dụng công thức 
$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

```python
n = len(actual_price)
square_errors = []

for i in range(n):
    error = actual_price[i] - predicted_price[i]
    square_error = error ** 2
    square_errors.append(square_error)

mse_car = sum(square_errors) / n

print("Mean Squared Error (MSE):", mse_car)
```

### Vẽ Scatter Plot

```python
plt.figure(figsize=(10, 8))
plt.scatter(car_age, actual_price, color='blue', label='Actual Price')
plt.scatter(car_age, predicted_price, color='red', marker='o', label='Predicted Price')
plt.plot(car_age, predicted_price, color='red', linestyle='--', alpha=0.5)
plt.title('Car Age vs Price (Actual vs Prediction)')
plt.xlabel('Car Age (years)')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

# Tính MSE cho điểm số học sinh

```python
study_hours = np.array([2, 4, 6, 8, 10, 12, 14, 16])
actual_score = np.array([50, 65, 70, 80, 85, 92, 95, 97])
predicted_score = np.array([50, 60, 75, 78, 90, 88, 98, 100])
```

```python
mse_score = mean_squared_error(actual_score, predicted_score)
print("Mean Squared Error (MSE):", mse_score)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(study_hours, actual_score, color='blue', label='Actual Score')
plt.scatter(study_hours, predicted_score, color='red', marker='o', label='Predicted Score')
plt.plot(study_hours, predicted_score, color='red', linestyle='--', alpha=0.5)
plt.title('Study Hours vs Score (Actual vs Prediction)')
plt.xlabel('Study Hours (hours)')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()
```

# Tính MSE cho số lượng khách hàng theo chi tiêu quảng cáo

```python
ad_spend = np.array([500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000])
actual_customers = np.array([50, 80, 120, 150, 200, 250, 300, 320, 350, 380])
predicted_customers = np.array([55, 85, 110, 145, 195, 240, 290, 315, 340, 375])
```

```python
mse_customer = mean_squared_error(actual_customers, predicted_customers)
print("Mean Squared Error (MSE):", mse_customer)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(ad_spend, actual_customers, color='blue', label='Actual Customers')
plt.scatter(ad_spend, predicted_customers, color='red', marker='o', label='Predicted Customers')
plt.plot(ad_spend, predicted_customers, color='red', linestyle='--', alpha=0.5)
plt.title('Ad Spend vs Customers (Actual vs Prediction)')
plt.xlabel('Ad Spend ($)')
plt.ylabel('Customers')
plt.legend()
plt.grid(True)
plt.show()
```
