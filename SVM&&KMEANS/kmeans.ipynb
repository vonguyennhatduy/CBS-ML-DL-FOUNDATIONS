### Cài đặt mô hình K-Means Clustering

```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
centers = np.array([
    [-5, -5],
    [-5, 5],
    [5, -5],
    [5, 5]
])

blobs, groups = make_blobs(n_samples=200, n_features=2, centers=centers, cluster_std=1.5)
plt.scatter(blobs[:, 0], blobs[:, 1], c='gray')
plt.show()
```

### K-Mean Clustering Algorithm

```python
# khởi tạo các centroid ngẫu nhiên 

color = ['r', 'b', 'g', 'y']
centroids = np.array([
    [-1, 5],
    [-3, 12],
    [-6, 5],
    [-11, 4]
])



for i in range(4):
    plt.scatter(centroids[i:i+1,0], centroids[i:i+1 ,1], c=color[i], s=250)

plt.scatter(blobs[:, 0], blobs[:, 1], c='gray')
plt.show()
```

```python

for _ in range(250):

    pos = []
    # tính khoảng cách các điểm dữ liễu đến centroids, gần centroid nào nhất sẽ thuộc về chung cụm với centroid đó 
    for j in range(blobs.shape[0]):
        dis = np.sum((blobs[j] - centroids) ** 2, axis=1)
        argmin = np.argmin(dis)
        pos.append(argmin)


    pos = np.array(pos)

    # tính giá trị trung bình các điểm dữ liệu chung cụm để lấy centroid mới
    for j in range(centroids.shape[0]):
        centroids[j] = blobs[pos == j, :].mean(axis = 0)
```

```python
color = ['r', 'b', 'g', 'y']
plt.scatter(blobs[:, 0], blobs[:, 1], c='gray')


for i in range(4):
    plt.scatter(centroids[i:i+1,0], centroids[i:i+1 ,1], c=color[i], s=250)


plt.show()
```
